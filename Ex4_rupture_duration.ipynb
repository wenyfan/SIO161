```python
import numpy as np
from matplotlib import pyplot as plt

```

```python
theta0 = np.arange(0,360,1)
theta = theta0/180*np.pi
costheta = np.cos(theta)

plt.plot(theta0, costheta, '-')
```

```python
L = 20
V = 6
Vr = 3
taud = L/V*(V/Vr-costheta)

plt.plot(theta0, taud, '-')
```

```python
from scipy import signal
```

```python
window = signal.boxcar(2)*2
plt.plot(window)

window2 = signal.boxcar(10)
plt.plot(window2)
```

```python
astf = np.convolve(window,window2)

plt.plot(astf)
```
