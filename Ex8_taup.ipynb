```python
from obspy.taup import TauPyModel
model = TauPyModel(model="iasp91")
print(model)
```

```python
arrivals = model.get_travel_times(source_depth_in_km=10,
                                  distance_in_degree=20)
print(arrivals)
```

```python
arrivals1 = model.get_travel_times(source_depth_in_km=10,
                                  distance_in_degree=70,
                                  phase_list=["P", "PP"])
print(arrivals1)
```

```python
arrivals2 = model.get_ray_paths(
    source_depth_in_km=500, distance_in_degree=130, phase_list=["ttbasic"])
ax = arrivals2.plot_rays()
```

```python
arrivals3 = model.get_ray_paths(source_depth_in_km=5,
                               distance_in_degree=4,
                               phase_list=["P","S"])
ax = arrivals3.plot_rays(plot_type="cartesian")
```

```python
from obspy.taup import plot_travel_times
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(9, 9))
ax = plot_travel_times(source_depth=10, phase_list=["P", "S", "PP"],
                       ax=ax, fig=fig, verbose=True)
```
